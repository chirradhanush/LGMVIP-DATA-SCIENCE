```python
import matplotlib.pyplot as plt
```

```python
from skimage import io
```

```python
import cv2
```

```python
from google.colab.patches import cv2_imshow
```

```python
image=io.imread("https://images.unsplash.com/photo-1564507592333-c60657eea523?ixlib=rb-4.0.3&ixid=MnwxMjA3fDB8MHxzZWFyY2h8Mnx8dGFqJTIwbWFoYWx8ZW58MHx8MHx8&w=1000&q=80")
```

```python
cv2_imshow(image)
```

```python
gray_filter = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
cv2_imshow(gray_filter)
```

```python
inverted_filter = cv2.bitwise_not(gray_filter)
```

```python
cv2_imshow(inverted_filter)
```

```python
image_smoothing = cv2.GaussianBlur(inverted_filter, (21, 21), sigmaX=0, sigmaY=0)
```

```python
cv2_imshow(image_smoothing)
```

```python
def pencil_sketch(x, y):
  return cv2.divide(x, 255 - y, scale=256)
```

```python
pencil_sketch = pencil_sketch(gray_filter, image_smoothing)
```

```python
cv2_imshow(pencil_sketch)
```
